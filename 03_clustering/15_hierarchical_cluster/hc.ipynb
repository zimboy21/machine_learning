```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values
# x = dataset.iloc[:, 1:].values
```

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers (observations)')
plt.ylabel('Distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5, metric = 'euclidean', linkage = 'ward')
y_clusters = hc.fit_predict(x)
```

```python
plt.scatter(x[y_clusters == 0, 0], x[y_clusters == 0, 1], s = 100)
plt.scatter(x[y_clusters == 1, 0], x[y_clusters == 1, 1], s = 100)
plt.scatter(x[y_clusters == 2, 0], x[y_clusters == 2, 1], s = 100)
plt.scatter(x[y_clusters == 3, 0], x[y_clusters == 3, 1], s = 100)
plt.scatter(x[y_clusters == 4, 0], x[y_clusters == 4, 1], s = 100)
plt.title('Show Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spend Index')
plt.show()
```
