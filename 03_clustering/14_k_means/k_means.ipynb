```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values
# x = dataset.iloc[:, 1:].values
```

```python
from sklearn.cluster import KMeans

wcss = []
for cluster in range(1, 11):
  k_means = KMeans(n_clusters = cluster, init = 'k-means++', n_init = 10,random_state = 42)
  k_means.fit(x)
  wcss.append(k_means.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
k_means = KMeans(n_clusters = 5, init = 'k-means++', n_init = 10, random_state = 42)
y_clusters = k_means.fit_predict(x)

print(y_clusters)
```

```python
plt.scatter(x[y_clusters == 0, 0], x[y_clusters == 0, 1], s = 100)
plt.scatter(x[y_clusters == 1, 0], x[y_clusters == 1, 1], s = 100)
plt.scatter(x[y_clusters == 2, 0], x[y_clusters == 2, 1], s = 100)
plt.scatter(x[y_clusters == 3, 0], x[y_clusters == 3, 1], s = 100)
plt.scatter(x[y_clusters == 4, 0], x[y_clusters == 4, 1], s = 100)
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], c = 'black', s = 200)
plt.title('Show Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spend Index')
plt.show()
```
