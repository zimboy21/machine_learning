```python
#IMPORTING LIBRARIES AND DATASET

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Position_Salaries.csv')
```

```python
#SPLIT DATASET INTO DEPENDENT AND INDEPENDENT VARIABLES

x = dataset.iloc[:, 1 :-1].values
y = dataset.iloc[:, -1].values
```

```python
#TRAIN SIMPLE LINEAR REGRESSION MODEL ON WHOLE DATASET

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
lin_reg_res = lin_reg.predict(x)
```

```python
#TRAIN POLYNOMIAL REGRESSION MODEL ON WHOLE DATASET

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_reg_for_poly = LinearRegression()
lin_reg_for_poly.fit(x_poly, y)
poly_reg_res = lin_reg_for_poly.predict(x_poly)
```

```python
#VISUALISE RESULTS OF LINEAR REGRESSION

plt.scatter(x, y, color='red')
plt.plot(x, lin_reg_res, color='blue')
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.show()
```

```python
#VISUALISE POLYNOMIAL REGRESSION RESULT

poly_res = lin_reg_for_poly.predict(poly_reg.fit_transform([[6.5]]))

x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(x, y, color='red')
plt.scatter(6.5, poly_res, color='green')       #Salary predicted for the regional manager
plt.plot(x, poly_reg_res, color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.show()
```

```python
#PREDICT SINGLE OBSERVATION

position_level = 6.5    #region manager for a few years

poly_res = lin_reg_for_poly.predict(poly_reg.fit_transform([[position_level]]))
print('Expected salary for regional manager calculated with polynomial regression: ', poly_res)
```
